# What Are Sequences?

In Python, a _sequence_ is any collection of ordered elements. By using a sequence, you can use a single variable name to reference and hold multiple data objects. 

There are many types of sequences in Python, but the three most common are:
* **List** The most versatile sequence types, lists can contain any object, in any combination. Lists are _mutable_, which means their contents can be changed or _mutated_. Elements can be removed, added, or reassigned. Lists are created and indexed using `[]`.
* **Tuple** A tuple is very similar to a list, except that the contents of a tuple are _immutable_, and therefore cannot be changed. Tuples are created using `()`, but are still indexed using `[]`.
* **String** A string is a special type of sequence that only stores characters. Strings are created using `""` or `''` quotes, but are indexed in the usual way via `[]`.

In this course, we will use all three types of sequences, so it's important to know the differences between them. However, there are some similarities between them. For all types of sequences, indexing starts at `0`. The length of a sequence can be found using `len(seq)`.
